```python
# What are the top 10 countries with the highest populations?

# first creating a dataframe sorted by population

most_populous_df = countries_df.sort_values('population', ascending=False)
```

```python
# dropping 'world' and continents/groups of countries from the column 'location' so it isn't included in the results

most_populous_df.drop(most_populous_df.index[most_populous_df['location'] == 'World'], inplace=True)
most_populous_df.drop(most_populous_df.index[most_populous_df['location'] == 'Africa'], inplace=True)
most_populous_df.drop(most_populous_df.index[most_populous_df['location'] == 'Asia'], inplace=True)
most_populous_df.drop(most_populous_df.index[most_populous_df['location'] == 'Europe'], inplace=True)
most_populous_df.drop(most_populous_df.index[most_populous_df['location'] == 'European Union'], inplace=True)
most_populous_df.drop(most_populous_df.index[most_populous_df['location'] == 'North America'], inplace=True)
most_populous_df.drop(most_populous_df.index[most_populous_df['location'] == 'South America'], inplace=True)
```

```python
# finally dropping duplicated countries in the location column

most_populous_df = most_populous_df.drop_duplicates(subset=['location'])
```

```python
# finally we have a list of the 10 countries in the world with the highest population

most_populous_df = most_populous_df['location'].head(10)
```
