```python
# importing image recognition library and numpy 

from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# storing an example image in the variable 'img'

img = Image.open('exampleimage.jpg')
```

```python
# type checking the image

type(img)
```

#### Loading an image using PIL is simply a 3D numpy array containing pixel intensities for RGB channels of the image
#### We can convert the image into an array

```python
# converting image into array

img_array = np.array(img)
```

```python
# displaying the image using matplotlib

plt.imshow(img);
```

```python
# now we can see the height and width of the image in RGB pixels

img_array.shape
```

```python
# we can turn off the axis and gridlines of the image and give it a title

plt.grid(False)
plt.title('My desktop background')
plt.axis('off')
plt.imshow(img);
```

```python
# now lets zoom in and display only a small fraction of the image

plt.grid(False)
plt.axis('off')
plt.imshow(img_array[300:400, 350:500]);
```
